```python
forcible_rape_alaska= crime_2016_df[(crime_2016_df['STNAME'] == 'ALASKA') & (crime_2016_df['OFFENSE'] == 20)]
forcible_rape_alaska
```

```python
sexual_violence_US= crime_2016_df[(crime_2016_df['OFFENSE'] == 170) | (crime_2016_df['OFFENSE'] == 20)]
sexual_violence_US
#grouped together by offence codes regarding all sexual violence
```

```python
female_perpetrators = sexual_violence_US.eval('Female_Sum = F0_9 + F10_12 + F13_14 + F15 + F16 + F17 + F18 + F19 + F20 + F21 + F22 + F23 + F24 + F25_29 + F30_34 + F35_39 + F40_44 + F45_49 + F50_54 + F55_59 + F60_64 + F65')
print(female_perpetrators)
```

```python
population_df=pd.read_csv('population_2016.csv')
population_df.head()
```

```python
males_merged = male_perpetrators_gb.merge(population_df, left_on='STNAME', right_on='Selected Geographies')
males_merged.head()
```

```python
males_merged['perpetrators_per_capita'] = males_merged['Male_Sum'] / males_merged['Male']
males_merged.head()
```

```python
males_merged['perpetrators_per_100k_people'] = males_merged['Male_Sum'] / males_merged['Male'] * 100000
males_merged.head(55)
```

```python
males_merged.plot.bar(x='STNAME', y='perpetrators_per_100k_people')
plt.tight_layout()
plt.ylim(ymin=0, ymax=110)
plt.xlabel('State')
plt.ylabel('Male Arrests per 100k People')
plt.legend().remove()
plt.title('Male Arrests for Sexual Violence by State, 2016')
```
