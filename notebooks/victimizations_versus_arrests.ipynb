```python
YEAR = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
NISVS = [7916000, 8616000, 8659000, 7957000, 7438000, 8421000, 9762000, 9232000] #contact sexual violence
NCVS = [269700, 243800, 346830, 300170, 284350, 431840, 323450, 393980] #rape/ sexual assault
UCR_arrests = [76914, 67727, 66739, 60449, 59895, 56897, 74695, 56139] #arrests regarding sexual violence
UCR_known = [84767, 83425, 84376, 82109, 84864, 91261, 96970, 99856] #known offenses of rape
```

```python
fig = plt.figure(figsize=(11, 5))
ax = fig.add_subplot()
plt.plot(YEAR, NISVS, label = "CDC Estimate", color = 'mediumvioletred')
plt.plot(YEAR, NCVS, label = "DOJ Estimate", color = 'dodgerblue')
plt.plot(YEAR, UCR_known, label = "FBI Known Offences", color = 'darkorange', linestyle='dashed')
plt.plot(YEAR, UCR_arrests, label = "FBI Known Arrests", color = 'teal', linestyle='dashdot')

ax.spines[['top', 'right', 'left']].set_visible(False)
ax.tick_params(bottom=False)
ax.grid(False)

plt.ylabel('ylabel', fontsize=14)
plt.yticks(fontsize=11)

plt.style.use('dark_background')

plt.title('Estimates of Sexual Violence from Government Agencies, 2010-2017', fontsize=16)
plt.xlabel('Year')
plt.ylabel('People (in ten millions)')
plt.ylim(ymin=0, ymax=10000000)
plt.legend()
plt.show()
```

This graph represents CDC estimates with a pink line, DOJ estimates with a blue line, incidents known to the 
FBI and law enforecement with an orange line, and arrests known to the FBI with a green line. It shows the change of these estimates overtime as well as comparing them to one another. The most striking thing about this graph is the substantial gap between the CDC's estimates and the DOJ and FBI estimates. The mean number of cases estimated by the CDC being significantly higher at 8,500,125, versus the 344,265, 64,932, and 88,454 case estimates from the DOJ, FBI arrests, and FBI known offences respectively.

These four graphs are representative of the trends over time of each of the different agency's estimates of sexual violence. The data is plotted and a linear regression model is fit to each set of points. From this, it is evident that the CDC and DOJ reports are positively trending, as well as the FBI's known incidents, but the FBI's known arrests are trending negatively. This means that while it is being estimated that cases of sexual violence are rising, arrests are falling.
